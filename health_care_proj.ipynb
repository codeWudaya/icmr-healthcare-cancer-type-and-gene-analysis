# F Test

# '''
The t-test works well when dealing with two groups , but sometimes we want to compare more than two groups at the same time. '''

#It tell whether two or more groups are similar or not based on their mean similarity and f-score.

#Intution of 1 way Anova which is tested using F Distribution

'''If there are three attributes a1,a2,a3 and classes C1,C2,C3 if the means
of the three class is far from each other than that is a good attribute,
feature so if class means of a1 are (0.5,3.0,5.6) and a2 is (2,2.1,2.2) then
a1 is a better feature This is formally measured by F Score and it is higher from a1 and a2'''

```python
#not cat data

non_cat_data = main_data.drop(['Unnamed: 0'], axis=1)
non_cat_data.head()
```

```python
data_f_test=main_data
```

```python
#function for f test

def f_test(data_f_test,gene):  
    df_anova = data_f_test[[gene,'Class']]
    grps = pd.unique(df_anova.Class.values)
    grps
    d_data = {grp:df_anova[gene][df_anova.Class == grp] for grp in grps}
    F, p = stats.f_oneway(d_data['LUAD'], d_data['PRAD'], d_data['BRCA'], d_data['KIRC'], d_data['COAD'])
    print("p_values:-",p)
    if p<0.05:
        print("reject null hypothesis")
    else:
        print("accept null hypothesis")
        
    return 
```

```python
features=main_data.drop(['Unnamed: 0'],axis=1)
```

```python
inputt =features.values
```

```python
output = pd.get_dummies(y_lda)
```

```python
#importing train_test_split

from sklearn.model_selection import train_test_split


xx_train, xx_test, yy_train, yy_test = train_test_split(inputt,output
                                                        , test_size = 0.2, random_state=42)
```

```python
#printing the shapes

print(xx_train.shape)
print(xx_test.shape)
print(yy_train.shape)
print(yy_test.shape)
```

```python
#importing neccesary modules

import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.optimizers import Adam,RMSprop
from tensorflow.keras.layers import Dense, BatchNormalization, Dropout
```

```python
model = Sequential()

#adding input layer
model.add(Dense(10000, input_dim=20531, activation='relu', kernel_initializer='he_uniform'))


#Normalize the data
model.add(BatchNormalization())
# model.add(Dropout(0.2, seed = 123))

#Add 1st hidden layer
model.add(Dense(5000, activation='relu'))

#Add 2nd hidden layer
model.add(Dense(2000, activation='relu'))

#Add 3rd hidden layer
model.add(Dense(1000, activation='relu'))

#Add 4th hidden layer
model.add(Dense(500, activation='relu'))

#Add 5th hidden layer
model.add(Dense(200, activation='relu'))

#Add 6th hidden layer
model.add(Dense(100, activation='relu'))

#Adding OUTPUT layer
model.add(Dense(5, activation='softmax'))
```

```python
# SGD
#Compile the model
model.compile(optimizer = RMSprop(learning_rate = 0.01), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
#printing summary
model.summary()
```

```python
#fitting the model
h = model.fit(xx_train,yy_train,          
          validation_data=(xx_test,yy_test),
          epochs=5,
          batch_size=32)
```

```python
#ploting accuracy graph

plt.plot(h.history['accuracy'],label='training accuracy')
plt.plot(h.history['val_accuracy'],label='testing accuracy')
plt.legend()
plt.show()
```
